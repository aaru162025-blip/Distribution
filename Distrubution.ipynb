#Distribution



1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use
BINOM.DIST)

     -  Treat the number of 3s as a binomial random variable with n=30 and p=1/6.
P(X=5)=C(30,5)(1/6)^5(5/6)^25 ≈ 0.19211 (about 19.21%).
Excel formula: =BINOM.DIST(5,30,1/6,FALSE)

2. Generate 100 values in Excel using the continuous uniform distribution RAND() and plot a histogram.
Describe the shape of the distribution.

      -  RAND() generates values uniformly between 0 and 1. The histogram should appear approximately flat (rectangular), with bars of roughly equal height. Minor differences are due to random sampling.

3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and
55 if the data follows a normal distribution?

      -  Mean=50, SD=5. The interval 45–55 corresponds to z=-1 to z=+1. About 68.27% of values lie within one standard deviation of the mean.

4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

      - Standardization converts data to a common scale using z=(x-μ)/σ. The transformed data have mean 0 and standard deviation 1. It is important because it allows comparison of variables with different units, helps detect outliers, and is widely used in machine learning and statistical analysis.

5.  What is Kurtosis and their type?

      -  Kurtosis measures the heaviness of the tails and peakedness of a distribution.
Types:
• Mesokurtic: normal distribution (excess kurtosis = 0).
• Leptokurtic: heavier tails and sharper peak (positive excess kurtosis).
• Platykurtic: lighter tails and flatter peak (negative excess kurtosis).

6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die

      - A fair die gives each outcome (1–6) an equal probability of 1/6. Since every outcome is equally likely, the discrete uniform distribution is the correct probability model.

7. Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

      -  Compute P(X≥8)=1−P(X≤7). By symmetry for n=15 and p=0.5, P(X≥8)=0.5.
      Excel formula: =1-BINOM.DIST(7,15,0.5,TRUE), which returns 0.5.

8. How does log transformation help in stabilizing variance and making data more normally distributed?

      -  Log transformation reduces right skewness, compresses large values more than small ones, stabilizes variance (reduces heteroscedasticity), and often makes data closer to a normal distribution, improving the performance of many statistical methods.
